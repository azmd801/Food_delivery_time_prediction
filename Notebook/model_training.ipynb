```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data\cleaned_data.csv')
data.head()
```

```python
# from eda we check that year and month has no significant importance in prdiction of time take so we will drop them and also this unnamed columns
data=data.drop(labels=['Unnamed: 0','Month','Year'],axis=1)
data.head()
```

```python
# segregating independent and depemdaent feature
## Independent and dependent features
X = data.drop(labels=['Time_taken (min)'],axis=1)
Y = data[['Time_taken (min)']]
```

```python
# observing independent variables
X.head()
```

```python
# observing dependable variable
Y.head()
```

```python
# Define which columns should be one hot-encoded and which should be scaled
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(exclude='object').columns

print('categorical features: \n',categorical_cols)
print('\nnumerical features: \n',numerical_cols)
```

### Creating the training pielines

```python
from sklearn.impute import SimpleImputer ## HAndling Missing Values
from sklearn.preprocessing import StandardScaler # HAndling Feature Scaling
from sklearn.preprocessing import OneHotEncoder # one hot encoding
## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
## Numerical Pipeline
num_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler())

    ]

)

# Categorigal Pipeline
cat_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('one_hot_encoder',OneHotEncoder()),
    ]

)

preprocessor=ColumnTransformer([
('num_pipeline',num_pipeline,numerical_cols),
('cat_pipeline',cat_pipeline,categorical_cols)
])
```

```python
## Train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=30)
```

```python
# placing the transformed data into a data frame
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

## Model traning

```python
## Model Training

from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
# creating a function to evalute the models with using multiple metric

import numpy as np
from typing import Tuple

def evaluate_model(true: np.ndarray, predicted: np.ndarray) -> Tuple[float, float, float]:
    """
    Calculates several metrics to evaluate the performance of a regression model.

    Parameters:
    true (array-like): The true values of the target variable.
    predicted (array-like): The predicted values of the target variable.

    Returns:
    tuple: A tuple containing the mean absolute error (MAE), root mean squared error (RMSE),
           and R-squared (R2) value for the model's performance.
    """
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
## Train multiple models

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}

trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)

    #Make Predictions
    y_pred=model.predict(X_test)

    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```

```python
# printing inthe interrcept and coeffiecints of all models

for model in models:
    print(f'{model} coefficients:\n {models[model].coef_}\n')
    print(f'{model} intercept:\n {models[model].intercept_}\n')
    
```
